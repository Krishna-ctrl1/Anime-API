```python
'''
Season name     Months
winter:         January, February, March
spring:         April, May, June
summer          July, August, September
fall            October, November, December
'''
seasons = [
    "winter",
    "spring",
    "summer",
    "fall"
]
```

```python
anime_list = []

def get_anime_season(year, season):
    try:
        response = requests.get(f'https://api.myanimelist.net/v2/anime/season/{year}/{season}?limit=4',
                                headers={'X-MAL-CLIENT-ID': '1f47d012b55d75270470bb84295e2bf4'})
        response.raise_for_status()  
        print(f'Status ({year}/{season}):', response.status_code)
        
        data = response.json().get('data', [])
        
        for anime in data:
            anime_id = anime['node']['id']
            try:
                response_details = requests.get(
                    f'https://api.myanimelist.net/v2/anime/{anime_id}?fields=id,title,start_date,end_date,synopsis,mean,rank,popularity,num_list_users,num_scoring_users,nsfw,created_at,updated_at,media_type,status,genres,my_list_status,num_episodes,start_season,broadcast,source,average_episode_duration,rating,background,studios,statistics',
                    headers={'X-MAL-CLIENT-ID': '1f47d012b55d75270470bb84295e2bf4'}
                )
                response_details.raise_for_status()
                anime_list.append(response_details.json())
            except requests.exceptions.RequestException as e:
                print(f"Error fetching details for anime ID {anime_id}: {e}")
        
        print(f'({year}/{season}) done!')
        print('---')
    
    except requests.exceptions.RequestException as e:
        print(f"Error fetching season data for {year} {season}: {e}")


os.makedirs('dataset', exist_ok=True)

for year in range(2000, 2024):
    for season in ['winter', 'spring', 'summer', 'fall']:
        get_anime_season(year, season)
        time.sleep(30)  


if anime_list:
    anime_df = pd.DataFrame(anime_list)
    anime_df.to_csv('dataset/anime.csv', index=False)
    print("CSV file created successfully!")
else:
    print("No data collected. CSV file not created.")
```
